```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
houston = pd.read_csv('scored/HB_HOUSTON.csv')
```

```python
oldi = 0

for i in range(houston.shape[0]//1000 + 1):
    fromVal = oldi
    toVal = i*1000
    
    fname= ("{}_{}".format(fromVal, toVal))
    currHouston = houston[fromVal:toVal]

    try:
        pie = currHouston[['price', 'predicted_price']].plot(figsize=(20,10))
        fig = pie.get_figure()
        fig.savefig("plots/HB_HOUSTON_{}".format(fname))
    except Exception as e:
        print(str(e))
        
    oldi = i*1000
```
